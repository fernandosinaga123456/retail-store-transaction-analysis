```python
import pandas as pd
```

```python
df = pd.read_excel('store.xlsx')
df.head()
```

Total Revenue

```python
revenue_per_category = (
    df.groupby('product_category')['total_amount']
    .sum()
    .reset_index()
    .sort_values(by='total_amount', ascending=False)
)

revenue_per_category
```

Tren Penjualan per Bulan

```python
df['month'] = df['transaction_time'].astype(str).str[:7]
monthly_sales = (
    df.groupby('month')['total_amount']
    .sum()
    .reset_index()
    .sort_values('month')
)

monthly_sales
```

```python
df['year'] = df['transaction_time'].astype(str).str[:4]
yearly_sales = (
    df.groupby('year')['total_amount']
    .sum()
    .reset_index()
    .sort_values('year')
)

yearly_sales
```

```python
top_5_stores = (
    df.groupby('store_id')['total_amount']
    .sum()
    .reset_index()
    .sort_values(by='total_amount', ascending=False)
    .head(5)
)

top_5_stores
```

```python
top_product_categories = (
    df.groupby('product_category')['total_amount']
    .sum()
    .reset_index()
    .sort_values(by='total_amount', ascending=False)
)

top_product_categories
```

```python
payment_preferences = (
    df['payment_method']
    .value_counts()
    .reset_index()
)

payment_preferences.columns = ['payment_method', 'total_transactions']
payment_preferences
```

```python
discount_analysis = (
    df.groupby('discount')
    .agg(
        total_revenue=('total_amount', 'sum'),
        total_quantity=('quantity', 'sum'),
        total_transactions=('transaction_id', 'count')
    )
    .reset_index()
    .sort_values('discount')
)

discount_analysis
```

```python
product_performance = (
    df.groupby('product_category')['quantity']
    .sum()
    .reset_index()
    .sort_values(by='quantity', ascending=False)
)

product_performance
```

```python
summary = {
    'Total Revenue': df['total_amount'].sum(),
    'Total Transactions': df['transaction_id'].nunique(),
    'Total Products': df['product_id'].nunique(),
    'Total Stores': df['store_id'].nunique()
}

pd.DataFrame(summary.items(), columns=['Metric', 'Value'])
```
